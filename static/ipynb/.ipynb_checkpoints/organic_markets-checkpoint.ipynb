```python
# Import Dependecies 
import requests
import json
import pandas as pd
```

## Whole Foods

```python
# https://maps.googleapis.com/maps/api/place/textsearch/xml?query=restaurants+in+Sydney&key=YOUR_API_KEY
url = "https://maps.googleapis.com/maps/api/place/textsearch/json?query=Whole%20Foods+in+Maryland&key="
query_url = f'{url}{api_key}'
query_url
```

```python
response = requests.get(query_url).json()
print(json.dumps(response, indent= 4, sort_keys= True))
```

```python
store_name= []
latitude=[]
longitude=[]
rating= []
address = []
price_level = []


error_count = 0

for market in range(len(response["results"])):
    try:
        response = requests.get(query_url).json()
        store_name.append(response['results'][market]['name'])
        rating.append(response['results'][market]['rating'])
        address.append(response['results'][market]['formatted_address'])
        latitude.append(response['results'][market]['geometry']['location']['lat'])
        longitude.append(response['results'][market]['geometry']['location']['lng'])
        
    except KeyError as e:
        error_count += 1
        print(e)
```

```python
whole_foods= {
    'Store Name': store_name,
    "Rating": rating,
    "Lat": latitude,
    "Long": longitude,
    "Address": address
}

# Create a data frame from dictionary
whole_foods = pd.DataFrame(whole_foods)
whole_foods
```

```python
# Save whole_foods data into a csv file
whole_foods.to_csv("whole_foods.csv", index = False)
```
